# Bioinformática

Función para verificar que sea una cadena valida

```python
nucleotidos = ["A", "C", "G", "T"]

def validateSeq(seq):
    tmpsec = seq.upper()
    for nuc in tmpsec:
        if nuc not in nucleotidos:
            return False
    return tmpsec

cadena = "ttaGaCaCGCCCCca"
print(validateSeq(cadena))
```

### Cadena de nucleótidos
Función para generar de manera aleatoria una cadena de nucleótidos

```python
import random

nucleotidos = ["A", "C", "G", "T"]
randomdna = ''.join([random.choice(nucleotidos)
               for nuc in range(10000)])

print(validateSeq(randomdna))
```

Función para contar la frecuencia de nucleotidos en una cadena

```python
def countNucFrequency(seq):
    tmpFreqDict = {"A": 0, "C": 0, "G": 0, "T": 0}
    for nuc in seq:
        tmpFreqDict[nuc] += 1
    return tmpFreqDict

result = countNucFrequency(randomdna)
print(result)
print("A, C, G, T: ", ' '.join([str(val) for key, val in result.items()]))
```

### Complemento de una cadena

Los nucleótidos A y T son complementarios entre sí, al igual que G y C. Al tener una cadena y un suministro de nucleótidos “flotantes libres”, uno puede imaginar la síntesis de una cadena complementaria en una cadena plantilla. El principio y el final de una cadena de ADN se denominan 5' (pronunciado "cinco primos") y 3' (pronunciado "tres primos"), respectivamente.

reverse_complement.png

Función para generar el complemento de una cadena de ADN

```python
dna_comp = {'A':'T', 'T':'A', 'C':'G', 'G':'C'}

def complement(seq):
    return ''.join([dna_comp[nuc] for nuc in seq])

print(randomdna)
print(complement(randomdna))
```

```python
print(f" randomdna + complement:\n {randomdna}")
print(f" {''.join(['|' for c in range(len(randomdna))])}")
print(f" {complement(randomdna)}")
```

### Complemento inverso de una cadena de ADN

Función para generar el complemento inverso de una cadena

```python
dna_revcomp = {'A':'T', 'T':'A', 'C':'G', 'G':'A'}

def reverse_complement(seq):
    return ''.join([dna_revcomp[nuc] for nuc in seq])[::-1]

print(randomdna)
print(reverse_complement(randomdna))
```

```python
print(f" randomdna + reverse complement:\n 5'{randomdna} 3'")
print(f"   {''.join(['|' for c in range(len(randomdna))])}")
print(f" 3'{reverse_complement(randomdna)} 5' \n")
```

mundo en colores

```python
def colored(seq):
    bcolors = {
        'A': '\033[92m',
        'C': '\033[94m',
        'G': '\033[93m',
        'T': '\033[91m',
        'U': '\033[91m',
        'reset': '\033[0;0m'
    }

    tmpStr = ""

    for nuc in seq:
        if nuc in bcolors:
            tmpStr += bcolors[nuc] + nuc
        else:
            tmpStr += bcolors['reset'] + nuc

    return tmpStr + '\033[0;0m'

print(colored(randomdna))
```

### GC content
Es el porcentaje de nucleótidos de guanina (G) y citosina (C) en una secuencia de ADN o ARN. El contenido de GC puede tener importantes implicaciones en la estructura, función, y evolución de los genomas, así como en técnicas de secuenciación y análisis genómico.

```python
def gc_content(seq):
    """GC Content in a DNA/RNA sequence"""
    return round(
        ((seq.count('C') + seq.count('G')) / len(seq) * 100), 6)

print("GC: ", gc_content(randomdna), "%")
```

analisis de GC por ventanas

```python
def gc_content_subsec(seq, k=20):
        """GC Content in a DNA/RNA sub-sequence length k. k=20 by default"""
        res = []
        for i in range(0, len(seq) - k + 1, k):
            subseq = seq[i:i + k]
            res.append(
                round((subseq.count('C') + subseq.count('G')) / len(subseq) * 100))
        return res

print(gc_content_subsec(randomdna, k=100))
```

### Transcripción
La transcripción es el proceso de hacer una copia de ARN de la secuencia de ADN de un gen. Esta copia, llamada ARN mensajero (ARNm), transporta la información proteica del gen codificada en el ADN.

Función para transcribir la cadena de DNA a RNA

```python
def transcription(seq):
    """DNA --> RNA transcription, replacing Thymine with Uracil"""
    return seq.replace("T", "U")

print("cadena RNA: ", transcription(randomdna))
```
